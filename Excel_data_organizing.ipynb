```python
import pandas as pd

input_path = "Harcama_Ozeti.xlsx"
df = pd.read_excel(input_path)

def categorize(description):
    description = description.lower()
    if "7036" in description:
        return "Market"
    elif "aypara" in description or "pegasus" in description:
        return "Seyahat"
    elif "google" in description or "spotify" in description or "paycell/tvplus" in description or "patreon" in description or "openaı" in description:
        return "Abonelik"
    elif "yemeksepeti" in description or "dominos" in description:
        return "Online Yemek"
    elif "ıyzıco*amazon" in description or "hepsiburada" in description or "trendyol" in description:
        return "Online Alışveriş"
    else:
        return "diğer"

df["Kategori"] = df["Açıklama"].apply(categorize)

df_filtered = df[["Tarih", "Kategori", "Tutar"]]
df_filtered.rename(columns={"Tutar": "Ücret"}, inplace=True)

output_path = "iki_aylik_harcamalar_kuruslu_duzenli.xlsx"
df_filtered.to_excel(output_path, index=False)

print(f"Created Excel file with categories added and filtered: {output_path}")
```
